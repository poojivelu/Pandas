```python
#Task 33: Create a column "OrderSize" with three levels: "Small" for Quantity less than 5: "Medium" for Quantity between 5 and 20: "Large" for Quantity greater than 20
rdf['OrderSize']=''
rdf.loc[rdf['Quantity']<5,'OrderSize']='Small'
rdf.loc[(rdf['Quantity']>5) & (rdf['Quantity']<20),'OrderSize']='Medium'
rdf.loc[rdf['Quantity']>20,'OrderSize']='Large'
rdf
```
